## 3.1 - Meta features <a class="anchor" id="bullet3x1"></a>

```python
from tf.core.helpers import deepSize
import re

features=TF.features
totalFeatures=0
totalMembers=0
totalSize=0
print (f'feature            members        size (bytes)')
for feature in features:
    if re.match(r"mm_+", feature):
        data = features[feature].data
        if data is None:
            continue
        totalFeatures+=1
        members = len(data)
        totalMembers+=members
        size = deepSize(data)
        totalSize +=size
        print (f'{feature:18}{members:8}{size:20}')
print (f'{'-'*46}\nTotal           {totalMembers:10}{totalSize:20} for {totalFeatures} features')
```

## 3.2 - Summary features <a class="anchor" id="bullet3x2"></a>
